```python
import librosa
```

```python
x, sr = librosa.load('audio/simpleLoop.wav')
```

```python
print(x.shape)
print(sr)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```

```python
import IPython.display as ipd
```

```python
ipd.Audio('audio/CongaGroove-mono.wav') # load a local WAV file
```

```python
import numpy
sr = 22050 # sample rate
T = 2.0    # seconds
t = numpy.linspace(0, T, int(T*sr), endpoint=False) # time variable
x = 0.5*numpy.sin(2*numpy.pi*440*t)                # pure sine wave at 440 Hz
ipd.Audio(x, rate=sr) # load a NumPy array
```

```python
librosa.output.write_wav('audio/tone_440.wav', x, sr)
```
